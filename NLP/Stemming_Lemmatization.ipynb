```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
import string


file = open("NLPdataEx3&4-data_in.txt","r")
raw = file.read()

modified_stops = stopwords.words('english') + list(string.punctuation)

word_tokens = set(nltk.word_tokenize(raw))
filtered_words = [word for word in word_tokens if word not in modified_stops]
print(filtered_words)
```

```python
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
sbs = SnowballStemmer('english')
wnl = WordNetLemmatizer()

stemmed_words = [sbs.stem(i) for i in filtered_words]
lemmatized_words = [wnl.lemmatize(i) for i in filtered_words]
isEqual = [stemmed_words[i].upper() == lemmatized_words[i].upper() for i in range(len(filtered_words))]

df = pd.DataFrame({'Actual':filtered_words,'Stemmed':stemmed_words,'Lemmatized':lemmatized_words,'IsEqual_IgnoreCase':isEqual})
df[['Actual','Stemmed','Lemmatized','IsEqual_IgnoreCase']]
# sum(df.IsEqual_IgnoreCase)

# print("{:15} {:15} {:15} {:2}".format("Actual", "Stemmed", "Lemmatized", "IsEqual"))
# print("="*60)
# for i in filtered_words:
#     stemmed_word = ps.stem(i)
#     lemmatized_word = wnl.lemmatize(i)
#     print("{:15} {:15} {:15} {}".format(i, stemmed_word, lemmatized_word,bool(int(stemmed_word.upper() == lemmatized_word.upper()))))
```
